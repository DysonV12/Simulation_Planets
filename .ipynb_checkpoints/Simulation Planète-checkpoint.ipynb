# Simulation du mouvement des planètes:
___

## 1° Initialisation des planètes:


```python
import numpy as np
import math
import random
import pygame
import time
import sys
SIZE_X = int(1920)
SIZE_Y = int(1080)
nbr_planet = 5
```

```python
class planet:
    def __init__(self,mass,x,y,vx,vy,diametre,Fx = 0, Fy = 0,moove = True):
        self.mass = mass
        self.x = x
        self.y = y
        self.vx = vx
        self.vy = vy
        self.diametre = mass//10
        self.Fx = Fx
        self.Fy = Fy
        self.moove = moove
```

 ## 2° Mouvement des planètes:
 

```python
G = 6.67*pow(10,-11)
ordre = pow(10,22)
```

```python
def initialisation():
    planets = []
    for i in range(nbr_planet):
        a = random.randint(0,4)
        planets.append(planet(random.randint(1,255),random.randint(0,1920),random.randint(0,1080),random.randint(200000,500000),random.randint(200000,500000),random.randint(5,10),moove = a!=1))
    return planets
```

```python
planets = initialisation()
```

F = G * (m1 * m2) / r^2

```python
def force():
    for i,p1 in enumerate(planets):
        p1.Fx = 0
        p1.Fy = 0
    for i,p1 in enumerate(planets):
        for j in range(i+1,nbr_planet):
            r = math.sqrt(pow((planets[i].x-planets[j].x),2) + pow((planets[i].y-planets[j].y),2))
            planets[i].Fx += ((planets[j].x-planets[i].x)/r*pow(10,8)) * G*(planets[i].mass*planets[j].mass*ordre**2)/pow(r*pow(10,8),2)
            planets[j].Fx += ((planets[i].x-planets[j].x)/r*pow(10,8)) * G*(planets[i].mass*planets[j].mass*ordre**2)/pow(r*pow(10,8),2)
            planets[i].Fy += ((planets[j].y-planets[i].y)/r*pow(10,8)) * G*(planets[i].mass*planets[j].mass*ordre**2)/pow(r*pow(10,8),2)
            planets[j].Fy += ((planets[i].y-planets[j].y)/r*pow(10,8)) * G*(planets[i].mass*planets[j].mass*ordre**2)/pow(r*pow(10,8),2)
#r est en 10 puissance 8 m et mass en 10 puissance 22
```

```python
force()
```

```python
def coord(t):
    for i in range(nbr_planet):
        if planets[i].moove == True:
            planets[i].x = (1/2*(planets[i].Fx/(planets[i].mass*ordre))*t**2+planets[i].vx*t+planets[i].x)%1920
            planets[i].y = (1/2*(planets[i].Fy/(planets[i].mass*ordre))*t**2+planets[i].vy*t+planets[i].y)%1080
            planets[i].vx = planets[i].Fx/(planets[i].mass*ordre)+planets[i].vx
            planets[i].vy = planets[i].Fy/(planets[i].mass*ordre)+planets[i].vy        
```

```python
def show(i):
    print(planets[i].x)
    print(planets[i].y)
```

```python
show(0)
```

#                  Affichage avec pygame

```python
def simu(screen):
    for i,p in enumerate(planets):
        x = p.x
        y = p.y
        pygame.draw.circle(screen, (0,128,p.mass),(x,y),p.diametre) 
        force()
        coord(0.00000005)
```

```python
pygame.init()
pygame.display.init()
screen = pygame.display.set_mode((1920, 1080))
pygame.display.set_caption("PLANET SIMULATOR")
clock = pygame.time.Clock()
while True:
    screen.fill((0,0,0))
    simu(screen)
    pygame.display.update()
    clock.tick(500)
    for event in pygame.event.get() :
        if event.type == pygame.QUIT :
            pygame.quit()
            pass
    
        
        
   
```
